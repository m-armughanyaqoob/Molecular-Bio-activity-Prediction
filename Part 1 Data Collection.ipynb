```python
import pandas as pd

# Read the CSV file into a DataFrame
df= pd.read_csv('/Users/armughan/Downloads/bioactivity_data_raw.csv')

print(df.head())
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
import pandas as pd
bioactivity_threshold = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
# Assuming df2_nr and bioactivity_threshold are already defined
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2_nr[selection]

bioactivity_threshold_df = pd.Series(bioactivity_threshold, name='bioactivity_threshold')

# Concatenate df3 and bioactivity_threshold_df along the columns
df3= pd.concat([df3, bioactivity_threshold_df], axis=1)

# Display the combined DataFrame
print(df3)
```

```python
path_to_downloads = '/Users/armughan/Downloads/bioactivity_datapreprocessed.csv'
df3.to_csv(path_to_downloads, index=False)
```

```python
import pandas as pd
df= pd.read_csv('/Users/armughan/Downloads/acetylcholinesterase_03_bioactivity_data_curated.csv')
df
```

```python
df_no_smiles = df.drop(columns='canonical_smiles')
```

```python
smiles = []

for i in df.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
df_clean_smiles = pd.concat([df_no_smiles,smiles], axis=1)
df_clean_smiles
```
